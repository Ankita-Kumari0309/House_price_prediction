### Used to identify and filter out non-numeric values (like '34.5Sq. Meter') during data cleaning.

### converted squarefeet columns to numeric

```python
def convert_sqft_to_num(x):
    tokens=x.split('-')
    if(len(tokens)==2):
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

### Removed outliers in price per square foot within each location to ensure consistency and improve model accuracy by filtering values beyond 1 standard deviation from the mean.

```python
# Handling price per location
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        st = np.std(subdf['price_per_sqft'])
        reduced_df = subdf[(subdf['price_per_sqft'] > (m - st)) & (subdf['price_per_sqft'] <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    
    return df_out
```

```python
df=remove_pps_outliers(df)
```

# Handled 2 BHK vs 3 BHK Price Inconsistencies
→ Removed 3 BHK properties that were priced lower per sqft than 2 BHKs in the same location.

```python
import matplotlib.pyplot as plt

def plot_scatter_chart(df, location):
    bhk2 = df[(df.location == location) & (df.bhk == 2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]
    
    plt.rcParams['figure.figsize'] = (10, 10)  

    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price, marker='+', color='red', label='3 BHK', s=50)

    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price per Square Feet")
    plt.title(location)
    plt.legend()
    plt.show()  
```

```python
plot_scatter_chart(df, "Rajaji Nagar")
```

#### Rajaji Nagar Price Analysis – Key Findings
Compared 2 BHK vs 3 BHK apartments using a scatter plot of total_sqft vs price_per_sqft.

Found that some 3 BHKs are priced higher per sqft than 2 BHKs, which is unusual.

These indicate outliers or incorrect pricing.

Will be removed in preprocessing to improve model accuracy.



```python
def remove_outliers_bhk(df):
    exclude_indices = np.array([])

    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df.price_per_sqft < stats['mean']].index.values
                )

    return df.drop(exclude_indices, axis='index')
```

```python
df = remove_outliers_bhk(df)
```

### Kcross validation -
K-Fold Cross Validation is a model validation technique used to evaluate the performance of a model by splitting the dataset into K parts (folds), training the model on K-1 folds, and testing it on the remaining fold. This process is repeated K times, and the average score gives a more reliable estimate of model performance.

1. I used ShuffleSplit (a variant of cross-validation) from sklearn.model_selection to randomly shuffle and split the data into 5 folds.

2. This technique allowed me to train the model on different subsets and validate it on unseen data, improving the reliability of performance evaluation.

3. I applied cross-validation to multiple models like Linear Regression, Lasso, and Decision Tree, using cross_val_score() to compute the R² scores across the folds.

4. It helped in selecting the best model based on consistent performance across different data split

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score
from sklearn.linear_model import LinearRegression

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)  

scores = cross_val_score(LinearRegression(), X, y, cv=cv)
print(scores)
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor

# ShuffleSplit for consistency across all models
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

# 1. Linear Regression
lr = LinearRegression()
params_lr = {
    # 'normalize': [True, False]  # Optional: deprecated in newer sklearn versions
}
grid_lr = GridSearchCV(estimator=lr, param_grid=params_lr, cv=cv, scoring='r2', n_jobs=-1)
grid_lr.fit(X, y)

# 2. Decision Tree Regressor
dt = DecisionTreeRegressor(random_state=0)
params_dt = {
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_dt = GridSearchCV(estimator=dt, param_grid=params_dt, cv=cv, scoring='r2', n_jobs=-1, verbose=1)
grid_dt.fit(X, y)

# 3. Lasso Regression
lasso = Lasso()
params_lasso = {
    'alpha': [0.01, 0.1, 1, 10, 100],
    'max_iter': [1000, 5000, 10000]
}
grid_lasso = GridSearchCV(estimator=lasso, param_grid=params_lasso, cv=cv, scoring='r2', n_jobs=-1, verbose=1)
grid_lasso.fit(X, y)

# 4. Compare models
print("\nModel Comparison Results:\n")
models = {
    'Linear Regression': grid_lr,
    'Decision Tree': grid_dt,
    'Lasso Regression': grid_lasso
}

for name, model in models.items():
    print(f"{name}:")
    print(f"  Best R² Score   : {model.best_score_:.4f}")
    print(f"  Best Parameters : {model.best_params_}")
    print("-" * 40)
```

```python
def predict_price(area_type, location, sqft, bath, bhk):
    import numpy as np

    x = np.zeros(len(X.columns))

    x[X.columns.get_loc("total_sqft")] = sqft
    x[X.columns.get_loc("bath")] = bath
    x[X.columns.get_loc("bhk")] = bhk
    x[X.columns.get_loc("ready_to_move")] = 1

    # Handle case-insensitive matches
    area_type = area_type.lower()
    location = location.lower()

    # Set one-hot encoding value for area_type
    if area_type in X.columns.str.lower():
        col_index = list(X.columns.str.lower()).index(area_type)
        x[col_index] = 1

    # Set one-hot encoding value for location
    if location in X.columns.str.lower():
        col_index = list(X.columns.str.lower()).index(location)
        x[col_index] = 1

    return mode.predict([x])[0]
```
